# Data Visualization IBM

This is the final project for the Data Visualization course given by IBM on coursera. Its a quick visualization excercise about the data provided below.

A survey was conducted to gauge an audience interest in different data science topics, namely:
 - Big Data (Spark / Hadoop)
 - Data Analysis / Statistics
 - Data Journalism
 - Data Visualization
 - Deep Learning
 - Machine Learning
    
The participants had three options for each topic: Very Interested, Somewhat interested, and Not interested. 2,233 respondents completed the survey.

The survey results have been saved in a csv file and can be accessed through this link: https://cocl.us/datascience_survey_data.

Lets load up the data. 

```python
#Lets begin by importing the necessary libraries.
import pandas as pd
import numpy as np 
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
#Load up the data.
data = pd.read_csv('Topic_Survey_Assignment.csv')

#Renames the first column of the data frame. 
data.rename(columns={'Unnamed: 0': 'Categories'},inplace=True)

#Sets the index to be the categories and deletes the extra column
data.set_index('Categories', inplace=True)
data.index.name = None
```

```python
#Lets sort the values to have it in decreasing order.
data.sort_values(by='Very interested', ascending=False, inplace=True)

#Divide the data by the number of participants in the survey
data = data/2233
```

```python
#Lets use Matplotlib to plot the results of each data science field interest.

ax = data.plot(kind='bar', width=0.8,color=('#5cb85c', '#5bc0de', '#d9534f'), figsize=(20,8),fontsize=14)
ax.set_title("Percentage of Respondents' interest in Data Science Areas", fontsize=16)

#Remove the border lines for the following sides.
ax.spines['left'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

#Remove Y axis
ax.yaxis.set_visible(False)

# Legend font size to 14
ax.legend(fontsize=14)

#Write the percentages on top of each bar
for p in ax.patches:
    ax.annotate("{:.2%}".format(p.get_height()),xy=(p.get_x()+0.02, p.get_height() + 0.01),fontsize=14)
```
